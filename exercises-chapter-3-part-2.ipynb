```python
# Exercise 3 - Create a function that receive list with 4 elements as the parameter, then, add 2 more 
# elements to the list and print it, calling the function

# Inputs and creating the list:
print('Fistly, we need to create a list with 4 elements to be the parameter.')
print('Please, insert four elements below.')
list1 = ['','','','']
for i in range (0,4):
    list1[i] = input('Insert the element ' + str(i + 1) + ': \n')

# Creating the function:
def function3(parameter1):
    print('Now, we need to add two more elements to these list: \n')
    elem5 = input('Insert the fifth element: \n')
    elem6 = input('Insert the sixth element: \n')
    parameter1.append(elem5)
    parameter1.append(elem6)
    print('Please, find below the final list:')
    print(parameter1)
    
# Calling the function:
function3(list1)
```

```python
# Exercise 4 - Create a function that receives a formal argument and a possible list of elements. Make two
# callings to the function, with only one element in the first call and 4 elements in the second call.

# Inputs:
formal_argument = ['string']
possible_list = ['a','b','c']

# Creating the function:
# - 1st argument is mandatory
# - 2nd argument can be a list of multiple elements (not a certain number)
def function4(formal_argument, *possible_list):
    formal_argument = list(formal_argument)
    for i in range(0, len(possible_list)):
        formal_argument.append(possible_list[i])
    print(formal_argument)
    print('')

# First call - only the formal argument:
function4(formal_argument)

#Second call - the formal argument and a possible list of elements:
function4(formal_argument, possible_list)
    
```

```python
# Exercise 5 - Create an anonymous function and attributes its return to a variable "sum_xy". The expression
# will receive 2 numbers as parameters and it will return the sum of them.

# Inputs:
print('Please, insert two integers to find their sum.')
int_1 = int(input('The first integer: \n'))
int_2 = int(input('The second integer: \n'))

# Defining the anonymous function:
sum_xy = lambda x,y: str(x + y)

# Calling and executing the funtion:
print('The sum is equal to ' + sum_xy(int_1, int_2))
```

```python
# Exercise 7 - Please, find below a list Cesius in temperature degrees.
# Create an anonymous function Fahrenheit to convert the temperature, using the lambda and map functions.
Celsius = [39.2, 36.5, 37.3, 37.8] # ---> Inputs

# From the theory we know that (C/5) = ((F-32)/9). So F = (9/5)*C + 32. Applytig them to the program:
Fahrenheit = list(map(lambda x: round((9/5)*x + 32, 2), Celsius))
print(Fahrenheit)
```
