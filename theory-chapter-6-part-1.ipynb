# Database

# DMBS (Database Management System)

# Relational Databases
#### It is a colection of tables with a relation among themselves.
#### Tables are connected using "Foreign keys".
#### Main Examples: ORACLE, Mycrosoft SQL Server, IBM DB2, PostgreSQL, SQLite, MySQL
#### Concepts:
    - Entities (Tables)
        Structure made with rows and collumns
    - Registers (Tuples) ---- Differente from the previous tuple definition ()
        Tuple/Register is each row of a table
    - Attributes:
        Attributes of a table consist in the collumns
    - Keys (Primary or Foreign)
        Useful to create relationship among different tables
        Primary Key (PK) is the main key of a table (only one primary key is allowed per table)
        Foreign Key (FK) are the collumns common in two or more different tables
    - Relationships among the entities (Tables)
        Follow the rules of relationships
    - Index
        It is an organized list with the data of the table. It is useful for the data reading.
    - Reference Integrity
        Addition, delete, or cascading update, etc.
    - Data Normalization
        Processes made to avoid unnecessary big tables.

# SQL (Structure Query Language)
#### SQL is the common language in working with related databases
#### SQL commands:
    DDL (Data Manipulation Language) - Useful for data development:
            - SELECT:   sql_select = 'select * from table_name'
            - UPDATE
            - DELETE
            - INSERT:  sql_insert = 'insert into table_name values (?, ?, ?)' 
    DML (Data Definition Language) - Useful for object definition:
            - CREATE:   sql_create = 'create table table_name','(column1_name column1_type primary key,'col2_name, column2_name column2_type, column3_name column3_type)'
            - ALTER
            - DROP
    DCL (Data Control Language) - Useful to determine privilege:
           - GRANT
           - REVOKE

# SQLite
#### SQLite is an SQL database engine useful for relational databases.
#### SQLite is a simplified SQL management program. It is included already in the conda.
#### Small (400 kb), simple, multiplataform, opensource, written in C language, no privilege.
#### It is advised for websites (<100k daily visits), Data Analysis (python, etc), IoT (smartwatch, smartphone, etc), CSV management, replacing Excel for SQLite, etc
#### Attention: SQL is not advised in data storage processes (website with large visits number), Server-Client applications, Big Data, etc

# Inserting data in a SQLite in Jupyter Notebook:

```python
# Data to be inserted (It is a list of tuples:)
data_input = [(1000, "Ciência de Dados", "Data Science"),
              (1001, "Big Data Fundamentos", 'Big Data'),
              (1002, 'Python Findamentos', 'Análise de Dados')]
```

```python
# Creating an instruction SQL to insert the registers/line/row:
sql_insert = 'insert into cursos values (?, ?, ?)'         # DDL - INSERT, table name = 'cursos'
```

```python
# Inserting the registers using SQL language:
for each_register in data_input:
    cursor_.execute(sql_insert, each_register)                
```

```python
# Saving the procedures with .commit():
connection.commit()
```

# Selecting a register of a SQLite:

```python
# Opening a query:
sql_select = 'select * from cursos'                  # DDL - SELECT, table name = 'cursos'
```

```python
# Selecting all the registers and recovering them:
cursor_.execute(sql_select)           # Selectig all registers in 'cursor_'
dados = cursor_.fetchall()            # Saving all registers in 'dados'
```

```python
# Displaying the registers:
for linha in dados:
    print('Curso ID: %d, Título: %s, Categoria %s \n' % linha)
```

# Inserting more registers to a SQLite Database:

```python
# Creating the registers to be added:
to_be_added = [(1003, 'Gestão de Dados com MongoDB','Big Data'),
               (1004, 'R Fundamentos', 'Análise de Dados')]
```

```python
# Inserting the new registers:
for registers in to_be_added:
    cursor_.execute(sql_insert, registers)
```

```python
# # Saving the procedures with .commit():
connection.commit()
```

```python
# Selecting all the new registers:
cursor_.execute(sql_select)
```

```python
# Recovering them:
new_dados = cursor_.fetchall()
```

```python
# Displaying the final database:
for registers in new_dados:
    print('Curso ID: %d, Título: %s, Categoria: %s \n' %registers)
```

```python
# Closing the connection:
connection.close()
```
